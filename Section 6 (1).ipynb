## On lit le fichier madrid_weather_data.csv dans une liste de lists

```python
weather_data = []

f = open("madrid_weather_2016.csv","r")
data = f.read()
rows = data.split("\n")

for row in rows:
    split_row = row.split(",")
    weather_data.append(split_row)
    
print(weather_data[0:5])

```

```python
count = 0

for item in weather_data:
    count += 1
    
print(count)
```
