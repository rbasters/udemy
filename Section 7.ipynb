## Qu'est-ce qu'un dictionnaire ?

```python
students = ["Thomas", "Matt", "Sarah", "Nina"]
scores = [16,15,17,14]

index = [0,1,2,3]
name = "Sarah"
score = 0

for i in index:
    if students[i]==name:
        score = scores[i]
print(score)
```

## Condition if / else

```python
planet_names = ["Mercure","Venus","Terre","Mars","Jupiter","Saturne","Neptune","Uranus"]
long_names=[]
short_names=[]
for planet in planet_names:
    if len(planet) > 5:
        long_names.append(planet)
    else:
        short_names.append(planet)
print(long_names)
print(short_names)
```

## Compter les éléments d'une liste et présenter le résultat dans un dictionnaire

```python
fruits=["citron","banane","pomme","pomme","pomme","banane","banane","pomme"]

fruit_count = {}

for item in fruits:
    if item in fruits:
        if item in fruit_count:
            fruit_count[item] += 1
        else:
            fruit_count[item] = 1
print(fruit_count)
```

Training comptage dictionnaire avec le fchier madrid:
 - Reprendre madrid_weather_2016.csv
 - créer un dictionnaire weather_count
 - Compter l'occurence de chaque climat et l'enregister
 - afficher à la fin

```python
weather_count = {}
weather_data = []

f = open("madrid_weather_2016.csv","r")
data = f.read()
temp_weather_data = data.split("\n")
temp_weather_data.pop(0) # on supprime l'en-tête

# split de la liste
for rows in temp_weather_data:
    split_data = rows.split(",")
    weather_data.append(split_data)

for item in weather_data:
        if item[1] in weather_count:
            weather_count[item[1]] += 1
        else:
            weather_count[item[1]] = 1

print(weather_count)
```
