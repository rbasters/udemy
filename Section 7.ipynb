# Les dictionnaires

## Qu'est-ce qu'un dictionnaire ?

```python
students = ["Thomas", "Matt", "Sarah", "Nina"]
scores = [16,15,17,14]

index = [0,1,2,3]
name = "Sarah"
score = 0

for i in index:
    if students[i]==name:
        score = scores[i]
print(score)
```

```python
scores = {} # dictionnaire vide
```

```python
scores["Marie"] = 15
scores["Nina"] += 2
scores["Thomas"] = 10

def present(nom):
    if nom in scores:
        print(nom,"est présent")
    else:
        print(nom,"n'est pas présent")

present("Matt")
present("Tom")
print(scores)
```
