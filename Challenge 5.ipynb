# # Ouverture du fichier et assignation à la variable <b>names</b>

```python
f = open("unisex_names.csv","r")
names = f.read()
names_list = names.split('\n')
print(names_list[0:10])
```

```python
names_data = []

for name in names_list:
    comma_list = name.split(",")
    names_data.append(comma_list)
    
print(names_data[0:5])
```

```python
numerical_list = []

for name in names_data:
    name_data = name[0]
    numerical_data = float(name[1])
    new_list=[name_data,numerical_data]
    numerical_list.append(new_list)
    
print(numerical_list[0:10])
```

```python
final_list = []

for entry in numerical_list:
    if entry[1] >= 1000:
        final_list.append(entry[0])
print(final_list[0:10])
```
